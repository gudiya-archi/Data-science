Creating DataFrame


```python
# using lists
student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data,columns=['id','marks','package'])
```

```python
# using dicts

student_dict = {
    'id':[100,90,120,80],
    'marks':[80,70,100,50],
    'package':[10,7,14,2]
}

pd.DataFrame(student_dict)
```
