```python
# import the library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plt.scatter simple function

x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data

df = pd.read_csv('/content/batter.csv')
df = df.head(50)
df

```

```python
plt.scatter(df['avg'],df['strike_rate'])
```

```python
plt.scatter(df['avg'],df['strike_rate'])
plt.title('Avg and SR anaiysis of top 50 Batsman')
```

```python
plt.scatter(df['avg'],df['strike_rate'])
plt.title('Avg and SR anaiysis of top 50 Batsman')

plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# we can change the color also

plt.scatter(df['avg'],df['strike_rate'],color='red')
plt.title('Avg and SR anaiysis of top 50 Batsman')

plt.xlabel('Average')
plt.ylabel('SR')
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='+')
plt.title('Avg and SR anaiysis of top 50 Batsman')

plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# stacked bar chart

plt.bar(df['batsman'],df['2017'])
plt.bar(df['batsman'],df['2016'],bottom=df['2017'])
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']))
```

```python
# stacked bar chart

plt.bar(df['batsman'],df['2017'] ,label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'], label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```
